```python
narratives = df['Consumer complaint narrative'].dropna().astype(str)
```

```python
import re 

def clean_text(text):
    text = text.lower() 
    text = re.sub(r'\bxxxx+\b','',text) #CFPB often masks names
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip() 

cleaned = narratives.apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    stop_words='english',
    ngram_range=(2,3),
    min_df=20
)

X = vectorizer.fit_transform(cleaned)
```

```python
freq = pd.DataFrame({
    'phrase': vectorizer.get_feature_names_out(),
    'count': X.sum(axis=0).A1
})

freq.sort_values('count', ascending=False).head(50)
```

```python
freq.to_csv("frequency.csv", index=False)
```
